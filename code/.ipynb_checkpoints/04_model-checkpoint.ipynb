### Import Libraries & Read Datasets

```python
#import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LassoCV, Lasso, RidgeCV, Ridge, ElasticNet, ElasticNetCV
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn import metrics
from sklearn.metrics import r2_score, mean_squared_error

%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
#import and read trian and test datasets into notebook
#train dataset
filepath_train = '../datasets/train_dummies.csv'

#test dataset
filepath_test = '../datasets/test_dummies.csv'

#read train_dummies csv
train = pd.read_csv(filepath_train, index_col="Id")

#read test_dummies csv
test = pd.read_csv(filepath_test, index_col="Id")
```

```python
#first few rows of train
print(train.shape)
train.head()
```

```python
#first few rows of test
print(test.shape)
test.head()
```

### Additional Cleaning of Data (Removing Outliers)

From the previous plot of Ground Living Area and Price in the EDA, we saw a few outliers worth examining to see if we can drop them.

```python
#plot regression plot of Ground Living Area with sales price
plt.figure(figsize=(15,10))
sns.regplot(data=train, x='Gr Liv Area',
            y='SalePrice', marker='o');
plt.title('Ground Living Area v. Sale Price', fontsize = 20)
plt.xlabel('Area(sqaure feet)', fontsize = 15);
plt.ylabel('Sale Price', fontsize = 15);
plt.savefig('outliers.jpg', dpi=300) 
```

We can see 2 extreme outliers in plot of ground living area and price. It is logical to remove these as well as price of the houses are simply too low for such a large area of living. We will remove those in order to not distort our model.

```python
#removing outliers that have area greater than 4000 square feet
train = train[train["Gr Liv Area"] < 4000]
```

```python
train.shape
```

```python
#Show the features with respect to SalePrice from high to low
train.corr()['SalePrice'].sort_values(ascending = False).head(20).to_frame()
```

### Preprocessing

##### Filling in missing columns

Before we start creating our matrix, we have to ensure there are same number of columns for both the train and test datasets for the model to work.

```python
#shape of train
print("Shape of train dataset: {}".format(train.shape))
#shape of test
print("Shape of test dataset: {}".format(test.shape))
```

```python
#obtain columns that are in test but not in train
columns_missing_train = list(set(test.columns) - set(train.columns))

#add missing columns to train
for cols in columns_missing_train: 
    train[cols] = 0
```

```python
#obtain columns that are in train but not in test
columns_missing_test = list(set(train.columns) - set(test.columns))

#add missing columns to test
for cols in columns_missing_test: 
    test[cols] = 0
```

```python
#shape of train
print("Shape of train dataset: {}".format(train.shape))
#shape of test
print("Shape of test dataset: {}".format(test.shape))
```

##### Create our features matrix (X) and target vector (y)

We will now create a matrix(X) that contains our highly correlated features that we want to fit into our model.

In addition, we will create our target(y) as the sales column we are trying to predict.

```python
#select only numeric features of train dataset
train_numeric = train.drop(columns = ['PID', 'SalePrice'], axis = 1)
features = list(train_numeric.columns)

#create our matrix(X)
X = train[features]

#create target salesprice that we want to predict
y = train['SalePrice']

#prepare X features for kaggle submssion
X_kaggle = test.drop(columns = ['PID', 'SalePrice'], axis = 1)
```

##### Train/test split

Split X and y variables into training and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

##### Scaling

We will now scale our variables in order to normalise the data. Features with numeric data like area is in square feet, where else feature of number of bedrooms is a discerete number and is unit less.  
These two features are on different scales, separated by a few orders of magnitude (square footage is in the thousands and number of bedrooms is in the single digits).

We will fit StandardScaler to X_train and use it to transform both X_train and X_test.


```python
ss = StandardScaler()
ss.fit(X_train)
#transfom training and testing
X_train_ss = ss.transform(X_train)
X_test_ss = ss.transform(X_test)
X_kaggle_ss = ss.transform(X_kaggle)
```

## Instantiate our models

We will be evaluating **LinearRegression**, **LassoCV**, **RidgeCV** and **ElasticNetCV** we thus we create instances for all 4 learning algorithms

```python
#LinearRegression algorithm
lr = LinearRegression()

#LassoCV algorithm
lasso_cv = LassoCV(n_alphas=200, max_iter = 10000)

#RidgeCV algorithm
ridge_cv = RidgeCV(alphas=np.logspace(0, 5, 200))

#ElasticNetCV algorithm
enet_cv = ElasticNetCV(l1_ratio=np.linspace(0.001, 1, 50), n_alphas = 100)
```

## Baseline: Linear Regression

Fit the model

```python
lr.fit(X_train_ss, y_train)
```

Obtain Train and Test score 

```python
#score for training data
print("Train data score: {}".format(lr.score(X_train_ss, y_train)))
      
#score for test data
print("Test data score: {}".format(lr.score(X_test_ss, y_test)))
```

Predictions

```python
#prediction of train data
predictions_train_linreg = lr.predict(X_train_ss)

#prediction of test data
predictions_test_linreg = lr.predict(X_test_ss)
```

Root Mean Square Error(RMSE)

```python
#RMSE of train
print("RMSE of train: {}".format(np.sqrt(metrics.mean_squared_error(y_train, predictions_train_linreg))))

#RMSE of test
print("RMSE of test: {}".format(np.sqrt(metrics.mean_squared_error(y_test, predictions_test_linreg))))
```

On first impression, the linear regression has a high R2 score for train data but very poor score for test data. Model is overfitted when dummies of nominal data are added, and thus linear regresison generally will not perform well with unseen data.

We will now use cross validation to compare scores for 4 models to see which provides the better results. 

## Cross validation

We use cross validation scores to estimate how well the 4 models work at train dataset level before testing on test dataset(unseen data)

```python
#scores for LinearRegression
lr_scores = cross_val_score(lr, X_train_ss, y_train, cv=5)
print("LinearRegression score: {}".format(lr_scores.mean()))

#scores for LassoCV
lasso_scores = cross_val_score(lasso_cv, X_train_ss, y_train, cv=5)
print("LassoCV score: {}".format(lasso_scores.mean()))

#scores for RidgeCV
ridge_scores = cross_val_score(ridge_cv, X_train_ss, y_train, cv=5)
print("RidgeCV score: {}".format(ridge_scores.mean()))

#scores for ElasticNetCV
enet_scores = cross_val_score(enet_cv, X_train_ss, y_train, cv=5)
print("ElasticNetCV score: {}".format(enet_scores.mean()))
```

```python
#scores for LinearRegression
lr_scores_test = cross_val_score(lr, X_test_ss, y_test, cv=5)
print("LinearRegression score: {}".format(lr_scores_test.mean()))

#scores for LassoCV
lasso_scores_test = cross_val_score(lasso_cv, X_test_ss, y_test, cv=5)
print("LassoCV score: {}".format(lasso_scores_test.mean()))

#scores for RidgeCV
ridge_scores_test = cross_val_score(ridge_cv, X_test_ss, y_test, cv=5)
print("RidgeCV score: {}".format(ridge_scores_test.mean()))

#scores for ElasticNetCV
enet_scores_test = cross_val_score(enet_cv, X_test_ss, y_test, cv=5)
print("ElasticNetCV score: {}".format(enet_scores_test.mean()))
```

## Model Fitting and Evaluation

###  Lasso Regression

As per the cross validation score, Lasso model seems to provide the best score amongst the 4 models 

Lasso regression helps to shrink the coeffcients of our features close to zero to reduce over-fitting. Thus less correlated features will be eliminated in the process 

We will use the Lasso model to fit the training data and evaluate both training and test datasets and to get our predictions.

Getting the optimised alpha for Lasso Model

```python
lasso_cv.fit(X_train_ss, y_train)
lasso_cv.alpha_
```

```python
lasso = Lasso(alpha=lasso_cv.alpha_)
#fit train data into linear regression model
lasso.fit(X_train_ss, y_train)

#score for training data
print("Train data score: {}".format(lasso.score(X_train_ss, y_train)))
      
#score for test data
print("Test data score: {}".format(lasso.score(X_test_ss, y_test)))
```

We use our predictions for both train and test data to find the Root Mean Square Error(RMSE) as metrics and plot the graph of predicted value against actual value to see the correlation.

```python
#prediction of train data
predictions_train_lasso = lasso.predict(X_train_ss)

#prediction of test data
predictions_test_lasso = lasso.predict(X_test_ss)
```

```python
#RMSE of train
print("RMSE of train: {}".format(np.sqrt(metrics.mean_squared_error(y_train, predictions_train_lasso))))

#RMSE of test
print("RMSE of test: {}".format(np.sqrt(metrics.mean_squared_error(y_test, predictions_test_lasso))))
```

```python
#plot graph of predicted vs actual saleprice
plt.figure(figsize=(20,15))

#scatterplot of predicted vs actual saleprice
plt.scatter(predictions_train_lasso, y_train)

#best fit line
plt.plot([0, np.max(y_train)],
         [0, np.max(y_train)],
         color = 'red')

plt.xlabel("Predicted SalePrice: $\hat{y}$", fontsize = 25)
plt.ylabel("Actual SalePrice: $y$", fontsize = 25)
plt.title('Predicted v. Actual Sale Price', fontsize = 30);
```

As per the plotted graph there a most of the prices are concentrated on the line of best fit, between 10000 to 25000 range with some obvious outliers.

We can also create a dataframe to look at the top 10 features that that affect the sale price.

```python
#create dataframe for lasso coefficents
lasso_coefs = pd.DataFrame({"Feature": features,
                          "coef": lasso.coef_,
                          "abs_coef": np.abs(lasso.coef_)
                           })

#sort lasso coefficients by abs_coef column
lasso_coefs[lasso_coefs["abs_coef"]>0].sort_values(by="abs_coef", ascending = False).head(10)
```

We can see that lasso does "feature selection" for us by providing the best features from "zero-ing" out coefficients of not so important features.   

The top features are also consistent to our EDA where by "Ground Living Area", "Overall Quality", "BsmtFin SF 1" and "Total Bsmt SF" have the largest absolute coefficients. Hence, overall quality and total area can be seen as the biggest factor in sale prices. 

We also note that the neighborhood of Northridge Heights had a large coefficient which could suggest location plays a part as well, and certain important amenities could be in this area that increases demand for a house in that area.  

```python
lasso_coefs[lasso_coefs["abs_coef"]==0]
```

We can see that Lasso model zeros out 117 features that are deemed not as important.

#### Summary for Lasso model
 
**Train** $R^2$ Score is at **92.3%** while **Test** $R^2$ Score is at **91.7%**  
**Train** RMSE is around **21946** while **Test** is around **22874**  

The model generally works well with both train and test data. However, the model fits better with our train data than our test data, which could be caused by residuals found when predicted price was plotted against actual price.

As both the $R^2$ Scores of train and test are similar, there is a relatively **high** bias and low variance. Therefore the lasso model maybe **underfit**. 

One way of improving $R^2$ scores and reducing RMSE is that we can try to remove outliers from the train data that creates "noise".

### Ridge Regression

The Ridge Model is best suited to deal with multicollinearity by shrinking the squares of the coefficient of each feature.

Obtain optimal alpha for ridge model

```python
#obtain optimal alphas of ridge
ridge_cv.fit(X_train_ss, y_train)
ridge_cv.alpha_
```

```python
ridge = Ridge(alpha = ridge_cv.alpha_)
```

R2 Score for Train and Test

##### Limitations

Predictions from our model could potentiall have been affected by missing values. Using logical imputation and using numerical imputation like mean, mode and median may be fast and cheap, but it definitely is not the best method in terms of accuracy. 

There is also multicollinearity within the data. For example, Overall Quality is has a high correlation to other features that is attributed to quality like kitchen and basement quality. Model will be more accurate and robust if features are to a large degree, independent with each other. 

There are many outliers still within the dataset that could cause distortion of results, these outliers maybe valid as well as they could be sold as at exorbitantly high prices on rare occasions. Hence the judgement of whether to keep or remove them is important. 

* Other Considerations to Explore:

  * Quality of a house is subjective to the potential house buyer
  * Upcoming infrastructure like roads and public transport may affect house prices too
